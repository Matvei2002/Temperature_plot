```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Чтение данных из CSV-файла с помощью Pandas
df = pd.read_csv('temperature_data.csv', encoding='utf-8')
```

```python
# Построение графика с отметками осей
plt.figure(figsize=(10, 6))
plt.plot(df['Время'], df['Датчик 1 (°C)'], label='Датчик 1')
plt.plot(df['Время'], df['Датчик 2 (°C)'], label='Датчик 2')

plt.xlabel('Время (секунды)')
plt.ylabel('Темпераутра (градусы)')
plt.title('Изменение температуры для датчиков 1 и 2')

# Установка пределов для осей X и Y
plt.xlim(0, 200)  # Ось X от 0 до 200
plt.ylim(0, 100)  # Ось Y от 0 до 100

# Добавление легенды с описанием кривых
plt.legend(loc='upper right')
plt.grid(True)

# Сохранение графика в формате png
plt.savefig('temperature_plot.png')
```
